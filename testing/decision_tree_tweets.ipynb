```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
import csv
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import tweepy
from nltk.tokenize import TweetTokenizer
import json

auth = tweepy.OAuthHandler('4VvS7ZE84ka02TqtxeNgf1gvq', 'AOSg6cV7pm9gZqvWxCSrctcTwNoo5ZnhgjA04KIvbRuo20w5It')
auth.set_access_token('2668704372-erQWOs8zPPdrW8HVPibsFcLTCB4aOTFiE5n7xbA', 'BYhhFwimMQyDt6JNavggmY696SfjVQSdnDxWkj7qjzi3A')
```

```python
api = tweepy.API(auth)

df = pd.read_csv('twitter_labelled/labeled_data.csv')

ids = list(df['ID'])
corpus = []
i = 0
Y = []
while i <= 3900:
    statuses = api.statuses_lookup(id_=ids[i: i+10])
    for ss in statuses:
        Y.append(df['label'][ids.index(ss.id)])
    
    corpus += [s.text for s in statuses]
    i += 100

vectorizer = CountVectorizer(min_df=10, stop_words='english', tokenizer=TweetTokenizer().tokenize)
X = vectorizer.fit_transform(corpus)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf2 = RandomForestClassifier(max_depth=100, random_state=0)
clf2 = clf2.fit(X_train, y_train)
clf3 = LinearSVC(random_state=0, tol=1e-5)
clf3 = clf3.fit(X_train, y_train)


res = clf.predict(X_test)
res2 = clf2.predict(X_test)
res3 = clf3.predict(X_test)
```

```python
acc = accuracy_score(y_test, res)
acc2 = accuracy_score(y_test, res2)
acc3 = accuracy_score(y_test, res3)
print(acc3)
```
