#### IMPORTS

```python
import json
from sklearn.model_selection import train_test_split
```

```python
with open('nytimestrial_annotations.json') as json_file:
    data = json.load(json_file)
```

```python
X, Y = [], []
i=0
for e in data["examples"]:
    if e["classifications"]:
        for i in range(len(e["classifications"])):
            if e["classifications"][i]["classname"] == "Bangladesh":
                continue
            else:
                if e["classifications"][i]["classname"] == "Not a flood":
                    
                    X.append(e["content"])
                    Y.append(0)
                else:
                    X.append(e["content"])
                    Y.append(1)
    else:
        X.append(e["content"])
        Y.append(0)
            
```

#### DATA PROCESSING

```python
import re
import numpy as np
import string
def remove_punct(text):
    text_nopunct = ''
    text_nopunct = re.sub('['+string.punctuation+']', '', text)
    return text_nopunct

X = [remove_punct(x) for x in X]
```

```python
import nltk
```

```python
# nltk.download('punkt')
# nltk.download('stopwords')
```

```python
from nltk import word_tokenize
tokens = [word_tokenize(sen) for sen in X]
```

```python
def lower_token(tokens): 
    return [w.lower() for w in tokens]    
    
lower_tokens = [lower_token(token) for token in tokens]
```

```python
from nltk.corpus import stopwords
stoplist = stopwords.words('english')
def removeStopWords(tokens): 
    return [word for word in tokens if word not in stoplist]
filtered_words = [removeStopWords(sen) for sen in lower_tokens]
X2 = [' '.join(sen) for sen in filtered_words]
```

```python
X_train,X_test,Y_train, Y_test = train_test_split(filtered_words, Y, test_size=0.1, random_state=42)
```

```python
all_training_words = [word for tokens in X_train for word in tokens]
training_sentence_lengths = [len(tokens) for tokens in X_train]
TRAINING_VOCAB = sorted(list(set(all_training_words)))
print("%s words total, with a vocabulary size of %s" % (len(all_training_words), len(TRAINING_VOCAB)))
```

```python
all_test_words = [word for tokens in X_test for word in tokens]
test_sentence_lengths = [len(tokens) for tokens in X_test]
TEST_VOCAB = sorted(list(set(all_test_words)))
print("%s words total, with a vocabulary size of %s" % (len(all_test_words), len(TEST_VOCAB)))
print("Max sentence length is %s" % max(test_sentence_lengths))
```
