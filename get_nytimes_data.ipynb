```python
def nytimes():
    for i in range(1980, 2021):
        print(i)
        year = str(i)
        link = 'https://api.nytimes.com/svc/search/v2/articlesearch.json?q=Bangladesh&q=flood&fq=pub_year:' + year +'&api-key=ICM207hTShMeUkRbW3PZAyqm3WSPX6L3&page=0'
        r = requests.get(link)
        res = r.json()["response"]["docs"]
        hits = r.json()["response"]["meta"]["hits"]
        pages = math.ceil(hits/10)
        print(pages)
        time.sleep(6)
        for j in range(1,pages):
            print(j)
            link = 'https://api.nytimes.com/svc/search/v2/articlesearch.json?q=Bangladesh&q=flood&fq=pub_year:' + year + '&api-key=ICM207hTShMeUkRbW3PZAyqm3WSPX6L3&page='+str(j)
            r = requests.get(link)
            res += r.json()["response"]["docs"]
            time.sleep(6)

        path = 'nytimes/' + str(i) + ".json"
        with open(path, 'w') as outfile:
            json.dump(res, outfile, indent=2)

nytimes()
```

```python
def save_links(year):
    fn = "nytimes/" + year + ".json"
    links = []
    with open(fn) as json_file:
        data = json.load(json_file)
        for p in data:
            links.append(p["web_url"])

    fn2 = "nytimes_article_links/" + year + ".txt"

    file1 = open(fn2, "w")
    for line in links:
        file1.write(line)
        file1.write("\n")
    file1.close()

for year in range(1980,2021):
    save_links(str(year))
```

```python
import os
import requests
from bs4 import BeautifulSoup
import json

def get_content(soup_obj):
    if not soup_obj: return None
    return soup_obj.get('content')

site_year = {}
for file in os.listdir("nytimes_article_links"):
    print(file) 
    sites = []
    if ".txt"in file:
        path = "nytimes_article_links/" + file
        f = open(path, "r")
        for line in f:
            sites.append(line.split('\n')[0])
        f.close()
        site_year[file] = sites
       
        
for key in site_year:
    sites = site_year[key]
    all_sites_data = []
    for site in sites:
        site_data = ""
        page = requests.get(site)
        soup = BeautifulSoup(page.text, 'html.parser')

        headline = soup.find('h1', itemprop='headline')
        if headline: site_data += headline.text + '\n'

        publishing_date = get_content(soup.find("meta", property="article:published"))
        if publishing_date: site_data += publishing_date + '\n'

        article = soup.find('section', itemprop='articleBody')
        if article:
            text_p = article.find_all('p')
            if text_p: article = '\n'.join([p.text for p in text_p])

        if article: site_data += article

        if site_data: all_sites_data.append({'text':site_data})
        
        print(".")
    
    f_name = "all_articles/" + key.split(".")[0] + ".json"
    print(f_name)
    with open(f_name, 'w') as outfile:
        json.dump(all_sites_data, outfile)
```
